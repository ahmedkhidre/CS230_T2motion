```python
from sentence_transformers import SentenceTransformer
import numpy as np
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
sentences = ["A man walks forward", "A man is walking straight", "A man is running", "A man is standing still"]

# I expect the embeddings of the first two sentences to be relatively close to each other, since they are saying the same thing
embeddings = model.encode(sentences)

print("Difference between _%s_ and _%s_: %.2f" % (sentences[0], sentences[1], np.linalg.norm(embeddings[0] - embeddings[1])))
```

```python
sentences = ["he is a king", "she is a queen", "he is a man", "she is a woman"]

# I expect the difference between embeddings of the first two sentences to be similar to the difference of the embeddings of the last two sentences, since in both cases the only difference is gender
embeddings = model.encode(sentences)

print("Actual difference: %.2f" % np.linalg.norm((embeddings[1] - embeddings[0]) - (embeddings[3] - embeddings[2])))
```
